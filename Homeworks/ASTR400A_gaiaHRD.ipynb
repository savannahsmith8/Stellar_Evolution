```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# chosen catalog item Messier 11 (Wild Duck cluster) which is an open cluster with a radius of 5 arc min
# assumed open clusters would have a large number of stars and would have younger stars than a globular
# when searching the Gaia database:
# 0.5 <= parallax <= 1.0
# ruwe <= 1.4 (eliminates, or at least limits, binary stars in the data)
# dataset contains 1768 stars
M11 = pd.read_csv("M11.csv")

# importing the file to read the MIST data
import read_mist_models
isocmd = read_mist_models.ISOCMD('MIST_iso_66eb8bf232a8f.iso.cmd')

# assigning variables to the iso data to add to HR diagram
age_ind = isocmd.age_index(8.5)
# number added to the end of the expression to shift the isochrone line to fit the HR diagram shape
mag_g = isocmd.isocmds[age_ind]['Gaia_G_DR2Rev'] + 10
bp_rp = isocmd.isocmds[age_ind]['Gaia_BP_DR2Rev'] - isocmd.isocmds[age_ind]['Gaia_RP_DR2Rev']

# plotting the isochrone line
plt.plot(bp_rp, mag_g, "m")
# plotting
plt.plot(M11.bp_rp, M11.phot_g_mean_mag, "oc", markersize = 0.75)
# inverting the y-axis
plt.gca().invert_yaxis()
# labeling plot
plt.title("HR Diagram for M 11, the Wild Duck Cluster")
# labeling the color index
plt.xlabel("BP_RP")
# Labeling the y-axis representing 'phot_g_mean_mag' in data
plt.ylabel("G-band magnitude")
plt.show()
```
