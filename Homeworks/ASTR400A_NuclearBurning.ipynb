```python
'''
For this assignment, I used the inputs that were suggested in D2L with a mass of 20 solar masses and the default stopping point (iron core). 
I used the trimmed_history data file from the MESA-Web file to plot each element's mass fraction (center_* columns) with respect to time. 
In certain cases, I plotted these mass fractions over a smaller time period to more accurately view the behavior of the mass fractions of the 
elements at the center of the star (i.e. after the main sequence). Comments at the start of every cell give an overview of the burning phases, 
their durations, and an explanation of what the plot displays.

important time periods:
estimated time before main sequence = 6.025e4 years
estimated time on the main sequence = 5119750 years ~ 5.12e6 years
estimated time of helium core burning = 332000 years ~ 3.32e5 years
estimated time of carbon core burning = 120000 years ~ 1.2e5 years
estimated time of oxygen core burning = 25000 years ~ 2.5e4 years
stopping condition was an iron core 
'''
```

```python
import mesa_reader as mr
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = mr.MesaData('C:\\Users\\savan\\Downloads\\MESA-Web_Job_10242435684\\MESA-Web_Job_10242435684\\trimmed_history.data')
```

```python
# overview of the main elements at the core of the star
# the main process represented here is the pp-chain where Hydrogen is burned into Helium
# once the Hydrogen has been burned, other elements grow in abundance and in the next plot, we can see a close-up of the post main sequence period
# at the very beginning of the time period, we can see oxygen slightly decrease, contributing to the CNO cycle

time = data.star_age

# main elements involved in stellar evolution that have characteristic curves

h1 = data.center_h1
he4 = data.center_he4
c12 = data.center_c12
o16 = data.center_o16

plt.title("mass fractions at the center of the star as a function of time")
plt.xlabel("time (years)")
plt.ylabel("mass fraction of abundance ($M_\u2609$)")

plt.plot(time, h1, 'm', label = "h1")
plt.plot(time, he4, 'r', label = "he4")
plt.plot(time, c12, 'y', label = "c12")
plt.plot(time, o16, 'g', label = "o16")

# estimated time before main sequence = 6.025e4 years

# this time period represents the period of the main sequence because during this time period, 
# the star burns its Hydrogen into Helium until it runs out of Hydrogen and therefore leaves the main sequence
plt.axvline(x = 6.025e4, label = "start of main sequence", ls = 'dashed', c = 'c')
plt.axvline(x = 5.180e6, label = "end of main sequence", ls = 'dashed')

# estimated time on the main sequence = 5119750 years ~ 5.12e6 years

plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left')
```

```python
# looking at the behavior of all the elements after the main sequence
# in the triple-alpha process, he4 core starts to burn and produce carbon 
# carbon fuses to increase oxygen abundance (and eventually burns closer to the stop time)

plt.title("mass fractions at the center of the star as a \nfunction of time (after main sequence)")
plt.xlabel("time (years)")
plt.ylabel("mass fraction of abundance ($M_\u2609$)")

ax = plt.gca()
ax.set_xlim(5e6, 5.6e6)

plt.plot(time, h1, 'm', label = "h1")
plt.plot(time, he4, 'r', label = "he4")
plt.plot(time, c12, 'y', label = "c12")
plt.plot(time, o16, 'g', label = "o16")

# the end of hydrogen burning represents the start of helium burning within the center of the star which can be seen below
plt.axvline(x = 5.180e6, label = "end of hydrogen burning", ls = 'dashed', c = 'm')
plt.axvline(x = 5.185e6, label = "start of helium burning", ls = 'dashed', c = 'r')

plt.axvline(x = 5.512e6, label = "start of oxygen burning", ls = 'dashed', c = 'g')
plt.axvline(x = 5.417e6, label = "start of carbon burning", ls = 'dashed', c = 'y')

# the end of helium burning can be seen towards the end of the time period
plt.axvline(x = 5.517e6, label = "end of helium burning", ls = 'dashed', c = 'c')

# end of burning for carbon and oxygen
plt.axvline(x = 5.537e6, label = "end of C/O burning", ls = 'dashed', c = 'k')

# estimated time of helium core burning = 332000 years ~ 3.32e5 years
# estimated time of carbon core burning = 120000 years ~ 1.2e5 years
# estimated time of oxygen core burning = 25000 years ~ 2.5e4 years

plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left')
```

```python
# all of these elements follow a similar pattern where they quickly peak at the very end of the time
# ne20 has a slight increase after the main sequence (from carbon fusion) but then follows the same trend as other elements
# most of these elements also increase as a result of carbon fusion 

ne20 = data.center_ne20
mg24 = data.center_mg24
si28 = data.center_si28
s32 = data.center_s32
ar36 = data.center_ar36
ca40 = data.center_ca40

plt.title("mass fractions at the center of the star as a function of time")
plt.xlabel("time (years)")
plt.ylabel("mass fraction of abundance ($M_\u2609$)")

plt.plot(time, ne20, 'c', label = "ne20")  # peaks at ~ 0.4
plt.plot(time, mg24, label = "mg24")       # peaks at ~ 0.15
plt.plot(time, si28, 'm', label = "si28")  # peaks at ~ 0.6
plt.plot(time, s32, 'r', label = "s32")    # peaks at ~ 0.4
plt.plot(time, ar36, 'y', label = "ar36")  # peaks at ~ 0.14
plt.plot(time, ca40, 'g', label = "ca40")  # peaks at ~ 0.13

plt.axvline(x = 6.025e4, label = "start of main sequence", ls = 'dashed', c = 'k')
plt.axvline(x = 5.180e6, label = "end of main sequence", ls = 'dashed')    # ne20 starts burning

plt.legend()
```

```python
# plotting all of the iron elements together that come from the Si group
# the most abundant is fe56, followed by fe54 (fe52 has very low abundance)
# makes sense that this would peak very high at the very end (essential 1.0 mass fraction) because the stopping condition for the 
# MESA-Web model was when the star formed an iron core

fe52 = data.center_fe52
fe54 = data.center_fe54
fe56 = data.center_fe56

plt.title("mass fractions at the center of the star as a function of time")
plt.xlabel("time (years)")
plt.ylabel("mass fraction of abundance ($M_\u2609$)")

plt.plot(time, fe56, label = "fe56")
plt.plot(time, fe54, 'c', label = "fe54")
plt.plot(time, fe52, 'g', label = "fe52") # peaks at ~ 0.0011 so the peak is not visible on the graph but it looks the same as fe54 and fe56

plt.axvline(x = 6.025e4, label = "start of main sequence", ls = 'dashed', c = 'y')
plt.axvline(x = 5.180e6, label = "end of main sequence", ls = 'dashed', c = 'k')

plt.legend()
```

```python
he3 = data.center_he3
ti44 = data.center_ti44
cr48 = data.center_cr48

# not included in main plots due to VERY low mass fraction

# This makes sense because although he3 is stable, it is used as an intermediate element in the pp chain

plt.title("mass fractions at the center of the star as a function of time")
plt.xlabel("time (years)")
plt.ylabel("mass fraction of abundance ($M_\u2609$)")

plt.plot(time, cr48, label = "cr48")       # peaks at ~ 3e-4 at the end of the time
plt.plot(time, ti44, 'y', label = "ti44")  # peaks at ~ 7e-5 at the end of the time
plt.plot(time, he3, 'm', label = "he3")    # starts at a mass fraction of ~ 3e-5 then drops to zero

plt.axvline(x = 6.025e4, label = "start of main sequence", ls = 'dashed', c = 'r')
plt.axvline(x = 5.180e6, label = "end of main sequence", ls = 'dashed', c = 'c')

plt.legend()
```

```python
# nitrogen used in the CNO cycle in very low abundance and can be seen to drop after the main sequence as expected

n14 = data.center_n14
ni56 = data.center_ni56

# not included in main plots due to low mass fraction

plt.title("mass fractions at the center of the star as a function of time")
plt.xlabel("time (years)")
plt.ylabel("mass fraction of abundance ($M_\u2609$)")

plt.plot(time, n14, 'g', label = "n14")
plt.plot(time, ni56, label = "ni56")

plt.axvline(x = 6.025e4, label = "start of main sequence", ls = 'dashed', c = 'c')
plt.axvline(x = 5.180e6, label = "end of main sequence", ls = 'dashed', c = 'k')

plt.legend()
```
