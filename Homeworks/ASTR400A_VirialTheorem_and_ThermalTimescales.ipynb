```python
import mesa_reader as mr
import numpy as np
import matplotlib.pyplot as plt

# reading the past MESA-Web profile files to plot Γ(m)
past_data_profile2 = mr.MesaData('C:\\Users\\Savan\\MESA-Web_Job_09252433840\\profile2.data')
past_data_profile8 = mr.MesaData('C:\\Users\\Savan\\MESA-Web_Job_09252433840\\profile8.data')

# reading the past MESA-Web data including the nuclear energy release and chemical evolution
past_data = mr.MesaData('C:\\Users\\Savan\\MESA-Web_Job_09252433840\\trimmed_history.data')
# reading the new MESA-Web data without the nuclear energy release and chemical evolution
new_data = mr.MesaData('C:\\Users\\Savan\\MESA-Web_Job_10072434724\\MESA-Web_Job_10072434724\\trimmed_history.data')
```

```python
# not part of the assignment of D2L but I had already plotted Γ(m)
m_p2 = past_data_profile2.mass
gamma_p2 = past_data_profile2.gamma1

m_p8 = past_data_profile8.mass
gamma_p8 = past_data_profile8.gamma1

plt.title("gamma at constant entropy as a function \nof mass of a $1M_\u2609$ star after 4.5 x $10^{9}$ years")
plt.xlabel("mass ($M_\u2609$)")
plt.ylabel("gamma (∂lnP/∂lnρ)")
plt.plot(m_p2, gamma_p2, 'm', label = "profile 2")
plt.plot(m_p8, gamma_p8, 'y', label = "profile 8")
plt.legend()
```

```python
logTeff_new = new_data.log_Teff
logL_new = new_data.log_L

plt.title("HR diagram for a star with with $1M_\u2609$ after 4.5 x $10^{9}$ years \nwithout nuclear energy release and chemical evolution")
plt.xlabel("log$_{10}$(T$_{eff}$)")
plt.ylabel("log$_{10}$(L)")
plt.gca().invert_xaxis()
plt.plot(logTeff_new, logL_new, 'g', label = "without NE release and CE")
plt.legend()
```

```python
logTeff_past = past_data.log_Teff
logL_past = past_data.log_L

plt.title("HR diagram for a star with with $1M_\u2609$ after 4.5 x $10^{9}$ years \nwith nuclear energy release and chemical evolution")
plt.xlabel("log$_{10}$(T$_{eff}$)")
plt.ylabel("log$_{10}$(L)")
plt.gca().invert_xaxis()
plt.plot(logTeff_past, logL_past, label = "with NE release and CE")
plt.legend()
```

```python
plt.title("HR diagram for a star with with $1M_\u2609$ after 4.5 x $10^{9}$ years")
plt.xlabel("log$_{10}$(T$_{eff}$)")
plt.ylabel("log$_{10}$(L)")
plt.gca().invert_xaxis()
plt.plot(logTeff_new, logL_new, 'g', label = "without NE release and CE")
plt.plot(logTeff_past, logL_past, label = "with NE release and CE")
plt.legend()
```
