```python
import mesa_reader as mr
import numpy as np
import matplotlib.pyplot as plt

# reading the MESA-Web data
data = mr.MesaData('C:\\Users\\Savan\\MESA-Web_Job_09252433840\\profile8.data')
```

```python
# assigning variables
m = data.mass
r = data.radius

# plotting
plt.title("mass versus radius relationship of a star \n with $1M_\u2609$ after 4.5 x $10^{9}$ years")
plt.ylabel("mass ($M_\u2609$)")
plt.xlabel("radius ($R_\u2609$)")
plt.plot(r, m, 'm', label = "m(r)")
plt.legend()
```

```python
logM = np.log10(m)
logR = np.log10(r)

plt.title("log$_{10}$ relationship between mass and radius")
plt.ylabel("log$_{10}$(M)")
plt.xlabel("log$_{10}$(R)")
plt.plot(logR, logM, 'm')
```

```python
L = data.luminosity

plt.title("luminosity versus radius relationship of a star \n with $1M_\u2609$ after 4.5 x $10^{9}$ years")
plt.ylabel("luminosity ($L_\u2609$)")
plt.xlabel("radius ($R_\u2609$)")
plt.plot(r, L, 'y', label = "L(r)")
plt.legend()
```

```python
logL = np.log10(L)

plt.title("log$_{10}$ relationship between luminosity and radius")
plt.ylabel("log$_{10}$(L)")
plt.xlabel("log$_{10}$(R)")
plt.plot(logR, logL, 'y')
```

```python
S = data.entropy

plt.title("entropy versus radius relationship of a star \n with $1M_\u2609$ after 4.5 x $10^{9}$ years")
plt.ylabel("entropy (ergs g$^{-1}$ K$^{-1}$)")   # not sure if these are the exact units that MESA-Web uses
plt.xlabel("radius ($R_\u2609$)")
plt.plot(r, S, 'g', label = "S(r)")
plt.legend()
```

```python
logS = np.log10(S)

plt.title("log$_{10}$ relationship between entropy and radius")
plt.ylabel("log$_{10}$(S)")
plt.xlabel("log$_{10}$(R)")
plt.plot(logR, logS, 'g')
```

```python
# plotting the above three plots as subplots to more clearly see the similar relationships with the radius of the star
fig, axs = plt.subplots(3)
plt.suptitle("variable relationships to the radius")
axs[0].plot(r, m, 'm')
axs[0].set_ylabel("mass \n($M_\u2609$)")
axs[1].plot(r, L, 'y')
axs[1].set_ylabel("luminosity \n($L_\u2609$)")
axs[2].plot(r, S, 'g')
axs[2].set_ylabel("entropy \n(ergs g$^{-1}$ K$^{-1}$)")

# labeling the x-axis for all three subplots
for ax in axs.flat:
    ax.set(xlabel = "radius ($R_\u2609$)")

# hides x values and ticks in between plots
for ax in axs.flat:
   ax.label_outer()
```

```python
# plotting the above three plots as subplots to more clearly see the similar relationships with the radius of the star
fig, axs = plt.subplots(2)
plt.suptitle("log$_{10}$ variable relationships to the radius")
axs[0].plot(logR, logM, 'm')
axs[0].set_ylabel("log$_{10}$(M)")
axs[1].plot(logR, logL, 'y')
axs[1].set_ylabel("log$_{10}$(L)")

# labeling the x-axis for all three subplots
for ax in axs.flat:
    ax.set(xlabel = "log$_{10}$(R)")

# hides x values and ticks in between plots
for ax in axs.flat:
   ax.label_outer()
```

```python
P = data.pressure

plt.title("pressure versus mass relationship of a star \n with $1M_\u2609$ after 4.5 x $10^{9}$ years")
plt.ylabel("pressure (dyn cm$^{-2}$)")
plt.xlabel("mass ($M_\u2609$)")
plt.plot(P, m, 'c', label = "P(m)")
plt.legend()
```

```python
plt.title("pressure versus radius relationship of a star \n with $1M_\u2609$ after 4.5 x $10^{9}$ years")
plt.ylabel("pressure (dyn cm$^{-2}$)")
plt.xlabel("radius ($R_\u2609$)")
plt.plot(P, r, label = "P(r)")
plt.legend()
```

```python
# plotting the above two plots as subplots to more clearly see the similar relationships with the pressure of the star
fig, axs = plt.subplots(1,2)
plt.suptitle("variable relationships to the pressure of the star")
axs[0].plot(P, m, 'c')
axs[0].set_xlabel("mass ($M_\u2609$)")
axs[1].plot(P, r)
axs[1].set_xlabel("radius ($R_\u2609$)")

# labeling the y-axis for both subplots
for ax in axs.flat:
    ax.set(ylabel = "pressure (dyn cm$^{-2}$)")

# hides y values and ticks in between plots
for ax in axs.flat:
   ax.label_outer()
```
